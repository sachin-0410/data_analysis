```python
# 1. Print the output series as the square of each number in the list.
#    First n natural number should appear in the index position
import pandas as pd
import numpy as np
my_list = [1, 2, 3, 4, 5]
pd.Series(my_list, index=list(np.arange(1, len(my_list)+1))).apply(lambda x: x**2)
```
